```python
headers = {
            'User-Agent': secrets.choice(userAgent)
        }
source = requests.get(f"https://covid-19.moh.gov.my/garis-panduan/garis-panduan-kkm", headers = headers)
source_status = str(source.status_code)
soup = BeautifulSoup(source.text, 'lxml')
annexes = soup.find_all('p')

info = []
for annex in annexes:

    if annex.a is not None:
        title = annex.a.text
        link = f"https://covid-19.moh.gov.my{annex.a.get('href')}"
        filesource = requests.get(link, headers = headers)
        fileName = title.replace('/', '-')
        fileName = fileName.strip()
        with open(f'{fileName}.pdf', "wb") as f:
            f.write(filesource.content)
        info.append({
            'Title': title,
            'Link': link
        })

df = pd.DataFrame(info)
df.to_csv(f'/KKM_COVID_Guidelines.csv')
```
