```python
import os.path
from os import path
dirName = 'KKM_media'
if path.exists(f'/content/{dirName}') == False:
    os.mkdir(f'/content/{dirName}')
```

```python
no_page = 66
phrase = 'database'
info = []

for n in range(1, no_page + 1):

    source_status = 100
    while source_status != "200":
        headers = {
            'User-Agent': secrets.choice(userAgent)
        }
        source = requests.get(f"https://www.moh.gov.my/index.php/database_stores/store_view/57?items=25&page={n}", headers = headers)
        source_status = str(source.status_code)
    soup = BeautifulSoup(source.text, 'lxml')
    titles = soup.find_all(lambda tag: tag.name == 'td' and not tag.attrs)
    titles = [title for title in titles if len(title.find_all()) == 0]
    titles = [title.text for title in titles if title.text != '']
    links = soup.find_all('a', {'title':'View', 'target':'_blank'})
    urls = []
    filenames = []
    filelinks = []
    for link in links:
        url = f'https://www.moh.gov.my{link.get("href")}'
        urls.append(url)
        try:
            file_source = requests.get(url, headers = headers)
            file_soup = BeautifulSoup(file_source.text, 'lxml')
            file_downloads = file_soup.find_all('th', {'colspan':'2'})
            for file_download in file_downloads:
                file_link = f'https://www.moh.gov.my{file_download.div.ul.li.a.get("href")}'
                file_name = file_download.div.ul.li.a.text
                if 'pdf' not in file_name:
                    file_name = file_name + '.pdf'

                with open(f'/content/{dirName}/{file_name}', "wb") as f:
                    try:
                        file_req = requests.get(file_link)
                        f.write(file_req.content)
                    except:
                        continue

                filelinks.append(file_link)
                filenames.append(file_name)
        except:
            file_link = ""
            file_name = ""

            filelinks.append(file_link)
            filenames.append(file_name)

    if len(titles) == len(urls) and len(urls) == len(filenames) and len(filenames) == len(filelinks):
        for i in range(len(titles)):
            info.append({
                'Title': titles[i],
                'URL': urls[i],
                'Filename': filenames[i],
                'Filelink': filelinks[i]
            })
    else:
        minEl = min([len(titles), len(urls), len(filenames), len(filelinks)])
        for i in range(minEl):
            info.append({
                'Title': titles[i],
                'URL': urls[i],
                'Filename': filenames[i],
                'Filelink': filelinks[i]
            })
    print(f"Currently done with page {n}")

df = pd.DataFrame(info)
df.to_csv(f'/content/{dirName}/KKM_media_statement.csv')
```

```python
source = requests.get("https://www.mkn.gov.my/web/ms/pelan-pemulihan-negara-fasa-4/").text
soup = BeautifulSoup(source, 'lxml')
topics = soup.find_all('a')
bodies = soup.find_all('tr')
```

```python
type = 'Four'
archivelst = []
count = 0
for body in bodies:
    if body.td != None:
        count = count + 1

print(f'{count} documents to retrieve')

fin = 0

for body in bodies:
    if body.td != None:
        tdlink, date = body.find_all('td')
        link = str(tdlink.a.get('href'))
        fileName = '_'.join(link.split('/')[7:])
        month = date.text.split('/')[1]
        year = date.text.split('/')[2]
        day = date.text.split('/')[0]
        mydict = {
            'fileName': fileName,
            'type': type,
            'day': day,
            'month': month,
            'year': year,
            'url': str(link)
        }
        archivelst.append(mydict)

        with open(fileName, "wb") as f:
            try:
                response = requests.get(link)
                time.sleep(1)
                f.write(response.content)
            except:
                try:
                    newstr = str(link)[1:]
                    response = requests.get(newstr)
                    time.sleep(1)
                    f.write(response.content)
                except:
                    pass
        fin = fin + 1
        print(f'{fin}/{count}')

df = pd.DataFrame(archivelst)
df.to_csv(f'National_recovery_plan_{type}.csv')
df.head()
```
