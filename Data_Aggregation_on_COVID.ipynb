```python
curr_time = time.time()

wd.get('https://legaldirectory.malaysianbar.org.my/v/search-result')
lawyer_tickbox = wd.find_element_by_id('searchtypelawyer')
firm_tickbox = wd.find_element_by_id('searchtypefirm')
wd.implicitly_wait(0.5)
firm_tickbox.click()

search_bar = wd.find_element_by_id('name')
wd.implicitly_wait(0.5)
search_bar.send_keys('*')

submit_btn = wd.find_element_by_id('submit')
wd.implicitly_wait(0.5)
submit_btn.click()


lst = []
start_page = 1801
while True:
    try:
        nextpg = Select(wd.find_element_by_id('lawyerpage1'))
        nextpg.select_by_value(f'{start_page}')
    except:
        continue
    break
maxpage = 2146
for pg_num in range(start_page + 1, maxpage + 1):
    wd.implicitly_wait(2)
    lawyer_names = []
    non_exact_loc = []
    curr_stat = []
    firm_name = []
    phone_num = []
    exact_loc = []
    email_address = []
    website = []

    print(f'Currently working on page {pg_num - 1}')
    while True:
        try:
            names = wd.find_elements_by_class_name('ftupc.ftbold')
            wd.implicitly_wait(0.1)
            for c in names:
                #print(c.text.replace("\'", "", 1).strip().lower())
                lawyer_names.append(c.text.replace("\'", "", 1).strip().lower())
        except:
            continue
        break

    while True:
        try:
            locations = wd.find_elements_by_class_name('tblCol2')
            wd.implicitly_wait(0.1)
            locations.pop(0), locations.pop(-1)
            for c in locations:
                non_exact_loc.append(c.text.lower().replace("\n", " "))
        except:
            continue
        break

    while True:
        try:
            statuses = wd.find_elements_by_class_name('tblCol4')
            wd.implicitly_wait(0.1)
            statuses.pop(0), statuses.pop(-1)
            for c in statuses:
                curr_stat.append(c.text.lower())
        except:
            continue
        break

    while True:
        try:
            count = 0
            firms = wd.find_elements_by_class_name('stblCol2')
            wd.implicitly_wait(0.1)
            for c in firms:
                count += 1
                if count == 1:
                    firm_name.append(c.text.strip().replace("\n", " "))
                elif count == 2:
                    phone_num.append(c.text.strip().replace("\n", " "))
                elif count == 6:
                    count = 0
                else:
                    continue
        except:
            continue
        break

    while True:
        try:
            addresses = wd.find_elements_by_xpath("//div[@class='ctblCol2']/span[@style='text-transform:uppercase;']")
            wd.implicitly_wait(0.1)
            for c in addresses:
                exact_loc.append(c.get_attribute('innerHTML'))
        except:
            continue
        break

    while True:
        try:
            teles = wd.find_elements_by_xpath("//div[@class='stblCol2' and @style='width:300px']")
            wd.implicitly_wait(0.1)
            pattern = r'"\s*(.*?)\s*"'
            count = 0
            for c in teles:
                count += 1
                ad = str(c.get_attribute('innerHTML')).replace("mailto:", "")
                match = re.search(pattern, ad)
                if count == 1:
                    if match is not None:
                        email_address.append(match.group(1))
                    else:
                        email_address.append("None")
                else:
                    if match is not None:
                        website.append(match.group(1))
                    else:
                        website.append("None")

                if count == 2:
                    count = 0
        except:
            continue
        break

    info_bulk = zip(lawyer_names, non_exact_loc, curr_stat, firm_name, phone_num, email_address, website, exact_loc)

    for _, info in enumerate(info_bulk):
        meta = {
            'Lawyer_Name':info[0],
            'Non_Exact_Loc':info[1],
            'Status':info[2],
            'Firm_Name':info[3],
            'Phone_Num':info[4],
            'Email':info[5],
            'Website':info[6],
            'Address':info[7]
        }
        lst.append(meta)

    sleep(random.uniform(0.1,0.3))
    while True:
        try:
            nextpg = Select(wd.find_element_by_id('lawyerpage1'))
            nextpg.select_by_value(f'{pg_num}')
        except:
            continue
        break

df = pd.DataFrame(lst)
df.to_csv('Malaysia_Lawyer_Information.csv')

end_time = time.time()
print(f'Used time:{end_time-curr_time}')
```

```python
count = 0
junior_counsel = []
data = []

for link in lst_of_links:
    wd_2.get(link)
    wd_2.implicitly_wait(2)
    Junior_Name = None
    Name_of_Chambers = None
    Address = None
    Tel_No = None
    Fax_No = None
    Email = None
    Website = None

    while True:
        try:
            name = wd_2.find_element_by_class_name('barrister-name')
            wd_2.implicitly_wait(0.2)
            Junior_Name = str(name.text).replace('\nBack','')
            #print(str(name.text).replace('\nBack',''))
        except:
            continue
        break

    while True:
        try:
            information = wd_2.find_elements_by_class_name('barrister-row')
            wd_2.implicitly_wait(0.3)
            for i in information:
                cur_info = str(i.text)
                if 'Name' in cur_info:
                    Name_of_Chambers = cur_info.replace('Name of Chambers:', '').replace('\n', ' ').strip()
                    #print(cur_info.replace('Name of Chambers:', '').replace('\n', ' ').strip())
                elif 'Address:' in cur_info:
                    Address = cur_info.replace('Address:', '').replace('\n', ' ').strip()
                    #print(cur_info.replace('Address:', '').replace('\n', ' ').strip())
                elif 'Tel.No:' in cur_info:
                    Tel_No = cur_info.replace('Tel.No:', '').replace('\n', ' ').strip()
                    #print(cur_info.replace('Tel.No:', '').replace('\n', ' ').strip())
                elif 'Fax.No:' in cur_info:
                    Fax_No = cur_info.replace('Fax.No:', '').replace('\n', ' ').strip()
                    #print(cur_info.replace('Fax.No:', '').replace('\n', ' ').strip())
                elif 'E-mail:' in cur_info:
                    Email = cur_info.replace('E-mail:', '').replace('\n', ' ').strip()
                    #print(cur_info.replace('E-mail:', '').replace('\n', ' ').strip())
                elif 'Website:' in cur_info:
                    Website = cur_info.replace('Website:', '').replace('\n', ' ').strip()
        except:
            continue
        break

    struct = {
        'Name': Junior_Name,
        'Name_of_Chambers': Name_of_Chambers,
        'Address': Address,
        'Tel_No': Tel_No,
        'Fax_No': Fax_No,
        'Email': Email,
        'Website': Website
    }

    data.append(struct)

    count += 1
    if count % 50 == 0:
        print(f'Working on the {count}th lawyer')

df = pd.DataFrame(data)
df.to_csv('HK_Senior_Counsel.csv')
```

```python
information = []
total = 0
delay = 5
start_time = time.time()
update_freq = 1000

while True:

    while True:
        try:
            #tr = wd.find_elements_by_xpath("//table[@id='Tbl_Search']/tbody/tr")
            tr = WebDriverWait(wd, delay).until(expected_conditions.visibility_of_element_located((By.XPATH, '//table[@id="Tbl_Search"]/tbody/tr')))
        except:
            continue
        break


    counter = 0
    bulk = []
    person = []
    while True:
        try:
            #md = tr.find_elements_by_xpath("//td/table/tbody/tr/td")
            md = WebDriverWait(tr, 10).until(expected_conditions.presence_of_all_elements_located((By.XPATH, "//td/table/tbody/tr/td")))
        except:
            continue
        break

    for i in md:
        #print(i.text, counter)

        if counter != 0 and counter != 6:
            person.append(i.text)

        counter+= 1

        if counter == 14:
            #print(person)
            bulk.append(tuple(person))
            person = []
            counter = 0

    name, reg, job, date, area, lp, typelaw, email, website, tel, fax, address = zip(*bulk)
    total += len(list(zip(*bulk))[0])
    for m in range(len(list(zip(*bulk))[0])):
        information.append({
            'Name': name[m],
            'Registration_Type': reg[m],
            'Job_Title': job[m],
            'Date_Admission': date[m],
            'Area': area[m],
            'Name_Law_Practice': lp[m],
            'Type_Law_Practice': typelaw[m],
            'Email': email[m],
            'Website': website[m],
            'Telephone': tel[m],
            'Fax': fax[m],
            'Address': address[m]
        })

    print(f'total of {total} has been recorded, {total/(time.time()-start_time)} records per sec on average')

    if total % update_freq == 0:
        df = pd.DataFrame(information, index=None)
        df.to_csv("Singapore_Lawyers.csv")

    try:
        next_btn = WebDriverWait(wd, delay).until(expected_conditions.element_to_be_clickable((By.XPATH, '//a[contains(@id, "next") and not (contains(@class, "disabledlink"))]')))
            #next_btn = wd.find_element_by_xpath('//a[contains(@id, "next") and not (contains(@class, "disabledlink"))]')
        wd.implicitly_wait(0.5)
        if next_btn is not None:
            while True:
                try:
                    next_btn.click()
                    wd.implicitly_wait(1)
                except:
                    continue
                break
        else:
            break
    except:
        break


df = pd.DataFrame(information, index=None)
df.to_csv("Singapore_Lawyers.csv")
```
